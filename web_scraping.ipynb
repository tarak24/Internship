## Q2

```python
page1=requests.get('https://www.imdb.com/chart/top/')
page1
```

```python
soup1=BeautifulSoup(page1.content)
soup1
```

```python
movies=soup1.find_all('td',class_='titleColumn')
movies
```

```python
movie=[]
year=[]
for i in movies:
    i=i.get_text().replace('\n',"")
    i=i.strip(" ")
    movie.append(i[:-6])
    year.append(i[-5:-1])
year
```

```python
Movie=[]
for j in movie:
    stri=''
    for k in j:
        if k.isalpha():
            stri=stri+k
    Movie.append(stri)
Movie
```

```python
Movie=Movie[:100]
year=year[:100]
```

```python
rating=[]
for i in soup1.find_all('td',class_='ratingColumn imdbRating'):
    i=i.get_text()
    i=i.replace('\n',"")
    rating.append(i)
rating
```

```python
rating=rating[:100]
```

```python
import pandas as pd
df=pd.DataFrame({'Name':Movie,'IMDb Rating':rating,'Year of release':year})
df
```

### i)

```python
page3=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page3
```

```python
soup3=BeautifulSoup(page3.content)
soup3
```

```python
Team=[]
Match=[]
Matches=[]
Points=[]
Ratings=[]
for i in soup3.find_all('span',class_='u-hide-phablet'):
    Team.append(i.get_text())
Team=Team[:10]
```

```python
x=soup3.find('td',class_='rankings-block__banner--matches')
x=x.get_text()
Matches.append(x)
y=soup3.find('td',class_='rankings-block__banner--points')
y=y.get_text()
Points.append(y)
z=soup3.find('td',class_='rankings-block__banner--rating u-text-right')
z=z.get_text()
z=z.replace('\n',"")
z=z.replace(" ","")
Ratings.append(z)
for i in soup3.find_all('td',class_='table-body__cell u-center-text'):
    Match.append(i.get_text())
Matches=Matches+Match[::2]
Points=Points+Match[1::2]
```

```python
for i in soup3.find_all('td',class_='table-body__cell u-text-right rating'):
    Ratings.append(i.get_text())
```

```python
Matches=Matches[:10]
Points=Points[:10]
Ratings=Ratings[:10]
```

```python
import pandas as pd
df2=pd.DataFrame({'team':Team,'matches':Matches,'points':Points,'ratings':Ratings})
df2
```
